```python
import random
import matplotlib.pyplot as plt
import statistics
```

```python
random.seed(42)
```

```python
# Populasi
populasi = [45, 50, 48, 52, 47, 55, 60, 43, 49, 51, 46, 54, 58, 62, 44, 53]
```

```python
#ulangi pengambilan sample
rata_rata_sample=[]
```

```python
jumlah_replikasi = 1000
ukuran_sample= 5
```

```python
for _ in range(jumlah_replikasi):
  sampel = random.choices(populasi, k=ukuran_sample)  # sampling with replacement
  rata_rata_sample.append(statistics.mean(sampel))
```

```python
print("Mean populasi:", statistics.mean(populasi))
print("Mean dari mean sampel:", statistics.mean(rata_rata_sample))
```

```python
plt.hist(rata_rata_sample, bins=30)
plt.title("Distribusi Mean Sampel")
plt.xlabel("Mean Sampel")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()
```
