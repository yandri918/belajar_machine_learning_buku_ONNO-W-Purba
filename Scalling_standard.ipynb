```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df =  pd.read_csv("feature_engineering_mahasiswa1.csv")
```

```python
fitur_numerik = ["jam_belajar", "kehadiran", "tugas"]
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(df[fitur_numerik])
```

```python
df_scaled = pd.DataFrame(scaled, columns=[f"{col}_std" for col in fitur_numerik])
```
