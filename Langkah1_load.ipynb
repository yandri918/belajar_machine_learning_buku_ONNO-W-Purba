```python
import pandas as pd
df = pd.read_csv("mini_project_regresi_rumah.csv")
```

```python
 print("=== DATA AWAL ===")
 print(df.head())
 print("\n=== SHAPE ===")
 print(df.shape)
 print("\n=== INFO ===")
 print(df.info())
 print("\n=== TIPE DATA ===")
 print(df.dtypes)
```

```python
print("=== SHAPE AWAL ===")
print(df.shape)
```

```python
# 2. Hapus duplikat
df = df.drop_duplicates()
```

```python
print("\n=== SETELAH HAPUS DUPLIKAT ===")
print(df.shape)
```

```python
# 3. Perbaiki tipe data
df["luas_rumah"] = pd.to_numeric(df["luas_rumah"], errors="coerce")
df["jumlah_kamar"] = pd.to_numeric(df["jumlah_kamar"], errors="coerce")
df["usia_rumah"] = pd.to_numeric(df["usia_rumah"], errors="coerce")
df["jarak_pusat_kota"] = pd.to_numeric(df["jarak_pusat_kota"], errors="coerce")
df["harga"] = pd.to_numeric(df["harga"], errors="coerce")
```

```python
print("\n=== MISSING VALUE ===")
print(df.isnull().sum())
```

```python
# 4. Isi missing value numerik dengan median
for kolom in ["luas_rumah", "jumlah_kamar", "usia_rumah", "jarak_pusat_kota", "harga"]:
  df[kolom] = df[kolom].fillna(df[kolom].median())
```

```python
# 5. Isi missing kategorikal dengan mode
df["kondisi_renovasi"] = df["kondisi_renovasi"].fillna(df["kondisi_renovasi"].mode()[0])
```

```python
# 6. Cek outlier sederhana pada harga dengan IQR
Q1 = df["harga"].quantile(0.25)
Q3 = df["harga"].quantile(0.75)
IQR = Q3 - Q1
```

```python
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR
```

```python
outlier = df[(df["harga"] < batas_bawah) | (df["harga"] > batas_atas)]
```

```python
print("\n=== OUTLIER HARGA ===")
print(outlier)
```
