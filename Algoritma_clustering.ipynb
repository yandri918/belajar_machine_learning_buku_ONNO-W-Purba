```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Dataset segmentasi pelanggan sederhana
```

```python
df = pd.DataFrame({ "belanja_bulanan": [1, 2, 2, 3, 8, 9, 10, 11],
                   "frekuensi_kunjungan": [1, 1, 2, 2, 8, 9, 9, 10] })
```

```python
df.columns
```

```python
X = df[["belanja_bulanan", "frekuensi_kunjungan"]]
```

```python
model = KMeans(n_clusters=2, random_state=42, n_init=10)
df["cluster"] = model.fit_predict(X)
```

```python
print("=== CLUSTERING ===")
print(df)
print("\nCentroid:")
print(model.cluster_centers_)
```
