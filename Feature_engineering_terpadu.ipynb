```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
print("=" * 80)
print("BAB 14 - FEATURE ENGINEERING DASAR")
print("=" * 80)
```

```python
# 1. LOAD DATA #
 #=========================================================
df = pd.read_csv("feature_engineering_mahasiswa1.csv")
```

```python
print("\n[1] DATA AWAL")
print(df.head())
```

```python
# 2. PILIH FITUR RELEVAN
```

```python
fitur = ["jam_belajar", "kehadiran", "tugas", "kota"]
```

```python
target_regresi = "nilai"
target_klasifikasi = "status_kelulusan"
```

```python
print("\n[2] FITUR TERPILIH:", fitur)
```

```python
#3. FEATURE TURUNAN
```

```python
df["rata_aktivitas"] = (df["jam_belajar"] + df["kehadiran"] + df["tugas"]) / 3
```

```python
df["rasio_tugas_per_jam"] = df["tugas"] / df["jam_belajar"]
df["interaksi_belajar_kehadiran"] = df["jam_belajar"] * df["kehadiran"]
```

```python
print("\n[3] FEATURE TURUNAN")
```

```python
print(df[[ "jam_belajar", "kehadiran", "tugas", "rata_aktivitas", "rasio_tugas_per_jam", "interaksi_belajar_kehadiran" ]].head())
```

```python
# 4. ENCODING TARGET DAN FITUR KATEGORI
```

```python
df["status_kelulusan_encoded"] = df["status_kelulusan "].map({"Tidak": 0, "Lulus": 1})
df = pd.get_dummies(df, columns=["kota"], dtype=int)
```

```python
print("\n[4] SETELAH ENCODING")
print(df.head())
```

```python
#5. SIAPKAN FITUR AKHIR
```

```python
fitur_akhir = [ "jam_belajar", "kehadiran", "tugas", "rata_aktivitas", "rasio_tugas_per_jam", "interaksi_belajar_kehadiran", "kota_Bandung", "kota_Depok", "kota_Jakarta", "kota_Surabaya" ]
```

```python
X = df[fitur_akhir]
y_reg = df["nilai"]
y_cls = df["status_kelulusan_encoded"]
```

```python
# 6. TRAIN-TEST SPLIT
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y_reg, test_size=0.2, random_state=42 )
```

```python
print("\n[5] TRAIN-TEST SPLIT")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
```

```python
# 7. SCALING YANG BENAR (ANTI DATA LEAKAGE)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("\n[6] SCALING SELESAI")
print("X_train_scaled shape:", X_train_scaled.shape)
print("X_test_scaled shape :", X_test_scaled.shape)
```
