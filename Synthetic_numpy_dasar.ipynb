```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
jam_belajar = np.random.randint(1, 11, size=20)
noise = np.random.randint(-5, 6, size=20)
nilai = 50 + (jam_belajar * 4) + noise
df = pd.DataFrame({ "jam_belajar": jam_belajar, "nilai": nilai })
print(df)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["jam_belajar"], df["nilai"])
plt.title("Synthetic Data dengan NumPy")
plt.xlabel("Jam Belajar")
plt.ylabel("Nilai")
plt.tight_layout()
plt.show()
```
