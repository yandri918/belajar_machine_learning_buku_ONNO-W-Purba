```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("eda_mahasiswa_bersih.csv")
```

```python
kolom_numerik = ["jam_belajar", "kehadiran", "tugas", "nilai"]
```

```python
corr_pearson = df[kolom_numerik].corr(method="pearson")
corr_spearman = df[kolom_numerik].corr(method="spearman")
```

```python
print("=== KORELASI PEARSON ===")
print(corr_pearson)
print("\n=== KORELASI SPEARMAN ===")
print(corr_spearman)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_pearson, annot=True, cmap="coolwarm")
plt.title("Heatmap Korelasi Pearson")
plt.tight_layout()
plt.show()
```
