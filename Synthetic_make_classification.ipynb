```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
# Membuat dataset klasifikasi sintetis
X, y = make_classification( n_samples=300,n_features=4,n_informative=3,n_redundant=0,n_classes=1,random_state=42 )
```

```python
df = pd.DataFrame(X, columns=["f1", "f2", "f3", "f4"])
df["kelas"] = y
```

```python
print(df["kelas"].value_counts())
```

```python
# Visualisasi
plt.figure(figsize=(8, 5))
plt.scatter(df["f1"], df["f2"], c=df["kelas"])
plt.title("Synthetic Classification Data")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.tight_layout()
plt.show()
```
