```python
def rekomendasi_algoritma(jenis_target, perlu_interpretasi=False, baseline_cepat=False, data_tabular_kompleks=False, butuh_segmentasi=False):
  if butuh_segmentasi:
    return "K-Means"
  if jenis_target == "angka":
    if perlu_interpretasi:
      return "Linear Regression"
    if data_tabular_kompleks:
      return "Random Forest / Gradient Boosting"
    return "Linear Regression"
  if jenis_target == "kelas":
    if perlu_interpretasi:
      return "Logistic Regression / Decision Tree"
    if baseline_cepat:
      return "KNN / Decision Tree"
    if data_tabular_kompleks:
      return "Random Forest / Gradient Boosting"
    return "Logistic Regression"
  if jenis_target == "tanpa_label":
    return "K-Means"
  return "Perlu definisi masalah lebih jelas"
```

```python
# Contoh penggunaan
print("Kasus 1:", rekomendasi_algoritma("angka"))
print("Kasus 2:", rekomendasi_algoritma("kelas", perlu_interpretasi=True))
print("Kasus 3:", rekomendasi_algoritma("kelas", baseline_cepat=True))
print("Kasus 4:", rekomendasi_algoritma("kelas", data_tabular_kompleks=True))
print("Kasus 5:", rekomendasi_algoritma("tanpa_label", butuh_segmentasi=True))
```
