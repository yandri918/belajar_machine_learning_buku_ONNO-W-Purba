```python
import pandas as pd
```

```python
df = pd.read_csv("feature_engineering_mahasiswa1.csv")
```

```python
# Feature turunan sederhana
df["rata_aktivitas"] = (df["jam_belajar"] + df["kehadiran"] + df["tugas"]) / 3
```

```python
df["rasio_tugas_per_jam"] = df["tugas"] / df["jam_belajar"]
```

```python
df["interaksi_belajar_kehadiran"] = df["jam_belajar"] * df["kehadiran"]
```

```python
# Binning jam belajar
df["kategori_jam_belajar"] = pd.cut( df["jam_belajar"], bins=[0, 4, 7, 12], labels=["Rendah", "Sedang", "Tinggi"] )
```

```python
print(df[[ "jam_belajar", "kehadiran", "tugas",
          "rata_aktivitas", "rasio_tugas_per_jam", "interaksi_belajar_kehadiran", "kategori_jam_belajar" ]].head(10))
```
