```python
def hitung_omzet_produk(jumlah, harga):
  return jumlah * harga
```

```python
def hitung_total_omzet(data):
  total = 0
  for item in data:
    total += item["omzet"]
  return total
```

```python
def hitung_rata_rata_omzet(data):
  if len(data) == 0:
    return 0
  return hitung_total_omzet(data) / len(data)
```

```python
def cari_produk_terlaris(data):
  if not data:
    return None
  produk_terlaris = data[0]
  for item in data:
    if item["jumlah"] > produk_terlaris["jumlah"]:
      produk_terlaris = item
  return produk_terlaris
```

```python
def cari_produk_omzet_tertinggi(data):
  if not data:
    return None
  produk_tertinggi = data[0]
  for item in data:
    if item["omzet"] > produk_tertinggi["omzet"]:
      produk_tertinggi = item
  return produk_tertinggi
```

```python
def cari_produk_omzet_terendah(data):
  if not data:
    return None
  produk_terendah = data[0]
  for item in data:
    if item["omzet"] < produk_terendah["omzet"]:
      produk_terendah = item
  return produk_terendah
```

```python
data_penjualan = []
```

```python
reader = pd.read_csv("penjualan_detail.csv")
```

```python
for index, row in reader.iterrows():
  produk = row["produk"]
  jumlah = int(row["jumlah"])
  harga = int(row["harga"])
  omzet = hitung_omzet_produk(jumlah, harga)
  data_penjualan.append({"produk": produk, "jumlah": jumlah, "harga": harga, "omzet": omzet})
```

```python
 data_penjualan.append({
     "produk": produk,
     "jumlah": jumlah,
    "harga": harga,
     "omzet": omzet         })
```

```python
total_omzet = hitung_total_omzet(data_penjualan)
rata_rata_omzet = hitung_rata_rata_omzet(data_penjualan)
produk_terlaris = cari_produk_terlaris(data_penjualan)
produk_tertinggi = cari_produk_omzet_tertinggi(data_penjualan)
produk_terendah = cari_produk_omzet_terendah(data_penjualan)
```

```python
if produk_terlaris is not None:
    print("Produk terlaris:", produk_terlaris["produk"])
else:
    print("Produk terlaris: Not found or function error.")

if produk_tertinggi is not None:
    print("Omzet tertinggi:", produk_tertinggi["produk"], produk_tertinggi["omzet"])
else:
    print("Omzet tertinggi: Not found or function error.")

if produk_terendah is not None:
    print("Omzet terendah:", produk_terendah["produk"], produk_terendah["omzet"])
else:
    print("Omzet terendah: Not found or function error.")
```

```python
print("Total omzet:", total_omzet)
print("Rata-rata omzet:", rata_rata_omzet)
```
