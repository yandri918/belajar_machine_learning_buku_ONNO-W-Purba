```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics  import mean_absolute_error, mean_squared_error, r2_score
import math
```

```python
# 1. Baca data bersih
df= pd.read_csv("eda_mahasiswa_bersih.csv")
```

```python
# 2. Pilih fitur dan target
X = df[["jam_belajar"]]
y = df["nilai"]
```

```python
# 3. Buat dan latih model
model = LinearRegression()
model.fit(X, y)
```

```python
# 4. Prediksi
y_pred = model.predict(X)
```

```python
# 5. Evaluasi
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
r2 = r2_score(y, y_pred)
```

```python
print("=== PARAMETER MODEL ===")
print("Intercept :", model.intercept_)
print("Koefisien :", model.coef_[0])
```

```python
print("\n=== EVALUASI MODEL ===")
print("MAE  :", mae)
print("MSE  :", mse)
print("RMSE :", rmse)
print("R²   :", r2)
```

```python
# 6. Tambahkan ke dataframe
df["prediksi_nilai"] = y_pred
df["residual"] = df["nilai"] - df["prediksi_nilai"]
```

```python
print("\n=== DATA DENGAN PREDIKSI ===")
print(df[["nama", "jam_belajar", "nilai", "prediksi_nilai", "residual"]])
```

```python
# 7. Visualisasi garis regresi
plt.figure(figsize=(8, 5))
plt.scatter(df["jam_belajar"], df["nilai"], label="Data aktual")
plt.plot(df["jam_belajar"], y_pred, label="Garis regresi")
plt.title("Regresi Linear Sederhana: Jam Belajar vs Nilai")
plt.xlabel("Jam Belajar")
```

```python
plt.ylabel("Nilai")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
