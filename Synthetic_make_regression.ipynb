```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Membuat dataset regresi sintetis
X, y = make_regression( n_samples=150, n_features=3, n_informative=2, noise=15, random_state=42 )
```

```python
# Ubah menjadi DataFrame agar mudah dibaca
df = pd.DataFrame({ "fitur_1": X[:, 0], "target": y })
```

```python
print(df.head())
```

```python
# Visualisasi
plt.figure(figsize=(8, 5))
plt.scatter(df["fitur_1"], df["target"])
plt.title("Synthetic Regression Data")
plt.xlabel("Fitur 1")
plt.ylabel("Target")
plt.tight_layout()
plt.show()
```
