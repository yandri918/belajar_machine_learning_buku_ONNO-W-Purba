```python
import pandas as pd
```

```python
df = pd.read_csv("statistik_mahasiswa.csv")
```

```python
print("=== LIMA BARIS PERTAMA ===")
print(df.head())
print("\n=== SHAPE ===")
print(df.shape)
print("\n=== TIPE DATA ===")
print(df.dtypes)
print("\n=== RINGKASAN AWAL ===")
print(df.describe())
```

```python
mean_nilai = df["nilai"].mean()
print("Mean nilai:", mean_nilai)
```

```python
median_nilai = df["nilai"].median()
print("Median nilai:", median_nilai)
```

```python
mode_nilai = df["nilai"].mode()[0]
print("Modus nilai:", mode_nilai)
```

```python
for kolom in ["nilai", "jam_belajar", "pengeluaran_bulanan", "kepuasan "]:
  print("=" * 50)
  print(f"Kolom: {kolom}")
  print("Mean   :", df[kolom].mean())
  print("Median :", df[kolom].median())
  print("Modus  :", df[kolom].mode                                                                                                                           ().tolist())
```

```python
var_nilai = df["nilai"].var()
print("Variansi nilai:", var_nilai)
```

```python
std_nilai = df["nilai"].std()
print("Standar deviasi nilai:", std_nilai)
```

```python
for kolom in ["nilai", "jam_belajar", "pengeluaran_bulanan", "kepuasan "]:
  print("=" * 50)
  print(f"Kolom: {kolom}")
  print(f"Minimum          : {df[kolom].min()}")
  print(f"Maksimum         : {df[kolom].max()}")
  print(f"Range            : {df[kolom].max() - df[kolom].min()}")
  print(f"Variansi         : {df[kolom].var()}")
  print(f"Standar deviasi  : {df[kolom].std()}")
```

```python
p25 = df["nilai"].quantile(0.25)
p50 = df["nilai"].quantile(0.50)
p75 = df["nilai"].quantile(0.75)
print("Persentil 25:", p25)
print("Persentil 50:", p50)
print("Persentil 75:", p75)
```

```python
Q1 = df["nilai"].quantile(0.25)
Q3 = df["nilai"].quantile(0.75)
IQR = Q3 - Q1
print("Q1 :", Q1)
print("Q3 :", Q3)
print("IQR:", IQR)
```

```python
batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR
```

```python
outlier = df[ (df["pengeluaran_bulanan"] < batas_bawah) | (df["pengeluaran_bulanan"] > batas_atas) ]
```

```python
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Batas bawah:", batas_bawah)
print("Batas atas :", batas_atas)
print("\nData yang terdeteksi sebagai outlier:")
print(outlier)
```
